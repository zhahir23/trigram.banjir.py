```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Added this line to download the missing resource
from nltk.tokenize import word_tokenize

def tokenize_text(text):
    return word_tokenize(text)

# Removed pd.read_csv as the file was not found and a df already exists
# The existing df from previous cell (RKJ_YMti3ms_) will be used.
# Assuming 'tweet' is the column containing text to be tokenized from the existing df.
df['tokenized_text'] = df['tweet'].apply(tokenize_text)
df['tokenized_text'].head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('indonesian')) # Using Indonesian stopwords
    return [word for word in tokens if word.lower() not in stop_words]

df['cleaned_text'] = df['tokenized_text'].apply(remove_stopwords)
display(df[['tweet', 'tokenized_text', 'cleaned_text']].head())
```

### membuat Sentiment Lexicons

```python
import pandas as pd

# Download lexicon files from a corrected URL
!wget https://raw.githubusercontent.com/masdevid/ID-Sentimen-Lexicon/main/data/positive.tsv -O /content/positive.tsv
!wget https://raw.githubusercontent.com/masdevid/ID-Sentimen-Lexicon/main/data/negative.tsv -O /content/negative.tsv

print("Content of positive.tsv:")
!head /content/positive.tsv # Using head to show only first few lines
print("\nContent of negative.tsv:")
!head /content/negative.tsv # Using head to show only first few lines

pos_lex = pd.read_csv(
  '/content/positive.tsv',
  sep='\t',
  engine='python',
  header=None,
  names=['word', 'weight']
)

neg_lex = pd.read_csv(
  '/content/negative.tsv',
  sep='\t',
  engine='python',
  header=None,
  names=['word', 'weight']
)

print("Positive Lexicon Columns:", pos_lex.columns)
display(pos_lex.head())
print("Negative Lexicon Columns:", neg_lex.columns)
display(neg_lex.head())

positive_words = dict(zip(pos_lex['word'], pos_lex['weight']))
negative_words = dict(zip(neg_lex['word'], neg_lex['weight']))
```

### kalkulasi Sentiment

```python
def sentiment_inset(text):
  score = 0

  for word in str(text).split():
    if word in positive_words:
      score += positive_words[word]
    elif word in negative_words:
      score += negative_words[word]
  return score

df['sentiment_score'] = df['cleaned_text'].apply(sentiment_inset)

def label(score):
  if score > 0:
    return 'Positif'
  elif score < 0:
    return 'Negatif'
  else:
    return 'Netral'

df['sentiment_label'] = df['sentiment_score'].apply(label)

display(df[['tweet', 'cleaned_text', 'sentiment_score', 'sentiment_label']].head())
```

## **Clean text**

```python
import re

def clean_text(text_column):
    text = text_column.lower()  # 1. Lowercase
    text = re.sub(r"http\S+", "", text) # 2. Hapus URL/Link
    text = re.sub(r"@\w+", "", text)  # 3. Hapus Mention (@username)
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # 4. Hapus Emoji dan Spesial Karakter (tapi SIMPAN angka & spasi) | ^a-zA-Z0-9\s = hapus selain huruf, angka, & spasi
    text = re.sub(r"\s+", " ", text).strip() # 5. Hapus spasi berlebih
    return text

df = df.drop_duplicates(subset=['tweet']).reset_index(drop=True)  # 6. Hapus data yang full_text-nya duplikat

df['clean_text'] = df['tweet'].apply(clean_text)
```

## **Trigram**

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Trigram
vectorizer = CountVectorizer(ngram_range=(3, 3), token_pattern=r'\b\w+\b')
X = vectorizer.fit_transform(df['clean_text'])

trigram_counts = X.toarray().sum(axis=0)
trigram_features = vectorizer.get_feature_names_out()
trigram_df = pd.DataFrame({
  'trigram': trigram_features,
  'count': trigram_counts
})

trigram_df = trigram_df.sort_values(by='count', ascending=False)
trigram_df.head(10)
```

```python
# Visualisasi Diagram
import matplotlib.pyplot as plt

top_trigram = trigram_df.head(10)
plt.figure(figsize=(10,5))

plt.barh(top_trigram['trigram'], top_trigram['count'])
plt.gca().invert_yaxis()
plt.title("Top 10 Trigram Paling Sering Muncul")
plt.xlabel("Frekuensi")
plt.show()
```
